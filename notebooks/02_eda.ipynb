```python
# imports

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load cleaned data

data = pd.read_csv("../data/cleaned_diabetes_data.csv")

data.head()
```

```python
# dataset overview

data.shape
data.info()
data.describe()
```

```python
# target variable

sns.countplot(x='diabetes_binary', data=data)
plt.title("Distribution of Diabetes Cases")
plt.show()

data['diabetes_binary'].value_counts(normalize=True)
```

```python
# Distribution of BMI

sns.histplot(data['bmi'], bins=30)
plt.title("BMI Distribution")
plt.show()
```

```python
# distirbution of age

sns.histplot(data['age'], bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
# BMI vs diabetes

sns.boxplot(x='diabetes_binary', y='bmi', data=data)
plt.title("BMI by Diabetes Status")
plt.show()
```

```python
# general health vs diabetes

sns.boxplot(x='diabetes_binary', y='genhlth', data=data)
plt.title("General Health vs Diabetes")
plt.show()
```

```python
# Correlation heatmap

# make sure outputs folder exists
os.makedirs("../outputs", exist_ok=True)

plt.figure(figsize=(12, 10))
sns.heatmap(data.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")

# SAVE the figure
plt.savefig("../outputs/correlation_heatmap.png", dpi=300, bbox_inches="tight")

# show it
plt.show()
```
